```
#2. Calculate the sum and average of first n natural numbers
#a. Using for loop
 
#code
 
n1=int(input("Enter the limit:"))
print("First",n1,"natural numbers:")
sum1=0
for i1 in range(1,n1+1):
  print(i1)
  sum1=sum1+i1
 
print("Sum=",sum1,"\nAverage=",sum1/n1)
```

```
 #b. Using while loop
 
#code
 
n2=int(input("Enter the limit:"))
print("First",n2,"natural numbers:")
sum2=0
i2=1 
while i2<=n2:
  print(i2)
  sum2=sum2+i2
  i2=i2+1
print("Sum=",sum2,"\nAverage=",sum2/n2)
```

```
#3. Accept list of 5 float numbers as input from user using loop and store values in a list
 
#code
 
lst1=[]
print("Enter 5 float numbers into the list :")
 
for il in range(0,5):
  lst1.insert(il,float(input()))
 
print(lst1)
```

```
#5. input a number, count and print the total number of digits in that number using while loop
 
#code
 
n=int(input("Enter a number:"))
tmp=n
sum=0
while n!=0:
  n=n//10
  sum=sum+1
 
print("Digits in",tmp,"is:",sum)
```

```
#6. Given two lists. Create a third list by picking an odd-index element from the first list and even index elements from second.
#listOne = [3, 6, 9, 12, 15, 18, 21]
#listTwo = [4, 8, 12, 16, 20, 24, 28]
 
#code
 
l1=[3, 6, 9, 12, 15, 18, 21]
l2=[4, 8, 12, 16, 20, 24, 28]
 
odd=l1[1::2]
even=l2[0::2]
l3=odd+even
print(l3)
```

```
#7. program to get the smallest number from a list.
 
#code
 
ls=[3,1,0.5,6,0.25,8]
nm=ls[0]
 
for i in ls:
  ns=i
  if nm>=ns:
    nm=ns
 
print ("Smallest number in list is:",nm)
```
